## Training

```python

df_train=df.drop(['building_id','damage_grade'],axis=1).copy()
target=df['damage_grade']
df_train.shape
```

```python
from pycaret.classification import *
passData=pd.concat([df_train,target],axis=1)
s=setup(data=passData,target='damage_grade')
```

```python
compare_models()
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame
# df = ...

# Iterate over each column
for col in df_train.columns:
    if df_train[col].dtype == 'object':  # Check if the column contains categorical data
        label_encoder = LabelEncoder()
        df_train[col] = label_encoder.fit_transform(df_train[col])

# Drop original categorical columns if needed
df_train.drop(df_train.select_dtypes(include='object').columns, axis=1, inplace=True)
```

```python
df_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier =RandomForestClassifier(n_estimators=1000,bootstrap=True)
classifier.fit(df_train,target)
```

## Predict

```python
test=pd.read_csv('test.csv')
```

```python
test.head()
test_final=test.drop(['building_id'],axis=1).copy()
```

```python
for col in test_final.columns:
    if test_final[col].dtype == 'object':  # Check if the column contains categorical data
        label_encoder = LabelEncoder()
        test_final[col] = label_encoder.fit_transform(test_final[col])

# Drop original categorical columns if needed
test_final.drop(test_final.select_dtypes(include='object').columns, axis=1, inplace=True)
```

```python
predicted = classifier.predict(test_final)
print(predicted)
```

## submit to kaggle

```python
prediction=pd.DataFrame(predicted)
sub_df=pd.read_csv('sample_submission.csv')
dataset=pd.concat([sub_df['building_id'],prediction],axis=1)
dataset.columns=['building_id','damage_grade']
dataset.to_csv('sample_submission.csv',index=False)
dataset
```
